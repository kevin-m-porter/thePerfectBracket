```python
from sportsipy.ncaab.schedule import Schedule
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from datetime import datetime
from dateutil import parser

TrankList = ['20110314_team_results.json','20120312_team_results.json','20130318_team_results.json',\
             '20140317_team_results.json','20150316_team_results.json','20160314_team_results.json',\
             '20170313_team_results.json','20180312_team_results.json','20190318_team_results.json',\
             '20210317_team_results.json']

tourneyGames = ['NCAA2011.csv','NCAA2012.csv','NCAA2013.csv','NCAA2014.csv','NCAA2015.csv','NCAA2016.csv',\
               'NCAA2017.csv','NCAA2018.csv','NCAA2019.csv','NCAA2021.csv']
Year = [2011,2012,2013,2014,2015,2016,2017,2018,2019,2021]


Translator = pd.read_csv(os.path.join('DATA','Translator.csv'),header=None)


for ind,(trank_file,games_file,year) in enumerate(zip(TrankList,tourneyGames,Year)):
    if year < 2017:
        continue
    print(f' Processing {year}')
    
    TRank = pd.read_json(os.path.join('DATA',trank_file), lines=False)
    Games = pd.read_csv(os.path.join('DATA',games_file),header=None)

    for indexs, game in Games[4:].iterrows(): # games 0-3 playin games

        winner= game[1]      
        loser = game[2]
        
        winner_trans = Translator[Translator[1].str.match(winner)]        

        datetime_object = datetime.strptime(game[0],  '%m/%d/%Y')
        
        winner_ref = winner_trans.iloc[0][0]
        
        schedule = Schedule(winner_ref,year=year)
        
        matched = False
        
        for gameref in schedule:
            if gameref.type =='NCAA' and datetime_object==parser.parse(gameref.date):
                matched= True
                #print(f'{game[0]}, Winner - {winner}, Loser - {loser}')
        
        if not matched:
            print(winner_ref)
            print(f'{game[0]}, Winner - {winner}, Loser - {loser} - No match??')

            for gameref in schedule:
                if gameref.type =='NCAA':
                    print(gameref.date)  # Prints the date the game was played
                    print(gameref.time)  # Prints whether the team won or lost
                    print(gameref.result)
                
                #print(f'Datetime {game[0]} {gameref.time}, Winner {winner}, Loser {loser}')
```
